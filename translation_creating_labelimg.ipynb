```python
dropped_parts = ['Instrumententafel','Lenkanlage','elektrisches System','Sitz hinten','elektronische Systeme',
                 'Getriebe','Motorkuehlung','Multimediasystem','Scheibenreinigung','Klimasystem','Gurtsystem hinten',
                'Heizungssystem','Vermessung','Airbagssystem','Unterlagen','Antriebsstrang','Pedalanlage',
                'Festpreis']
dropped_parts = [x.lower() for x in dropped_parts] 
def check_part(part):
    
    try:
        part = part.lower()
        if part.split('-')[0].strip() in (dropped_parts):
            return True
        else:
            return False
    except:
        print(part)


df['del'] = df.Location.apply(lambda x: check_part(x))
print(df.shape)
df = df.loc[df['del'] == False]
df.drop(['del'], axis=1, inplace=True)
print(df.shape)
```

```python
from tqdm import tqdm
de_eng = {}
for idx,row in tqdm(translation.iterrows()):
    
    if row.part_de not in de_eng.keys() and row.part_de != 'NAN':
        de_eng[row.part_de] = row.part_en
    else:
        pass
    if row.damage_de not in de_eng.keys() and row.damage_de != 'NAN':
        de_eng[row.damage_de] = row.damage_en
    else:
        pass
    if row.part_p1 not in de_eng.keys() and row.part_p1 != 'NAN':
        de_eng[row.part_p1] = row.part_p1_en
    else:
        pass
    if row.part_p2 not in de_eng.keys() and row.part_p2 != 'NAN':
        de_eng[row.part_p2] = row.part_p2_en
    else:
        pass
    if row.action not in de_eng.keys() and row.action != 'NAN':
        de_eng[row.action] = row.action_en
    else:
        pass
de_eng
```

```python
de_eng['06/19'] ='06/19'

def hotfix(x):
    try:
        return de_eng[x]
    except:
        return x
```

```python
df["damage_en"] = df.Damage.map(lambda x: de_eng[x])
df["action_en"] = df["Repair Action"].map(lambda x: de_eng[x])
df["part_en"] = df['Location'].map(lambda x: de_eng[x])
df["part_p1_en"] = df['Specific Location'].map(lambda x: de_eng[x])
df["part_p2_en"] = df['Detailed Location'].map(lambda x: hotfix(x))

```

```python
import os
chunks = 10
workers = ["w1","w2","w3","w4","w5"]
batch_root_dir = './person/'
csv_root_dir = './person/'
ann_root_dit = './person/'
damage_dict_batch_csv = dict()
idx = 0

for damage in damage_keys:
    df_dmg = df[df['Damage']==damage]
    try:
        samples= df_dmg.sample(3500)
    except:
        print(f'for {damage} sampling only: ',len(df_dmg))
        samples = df_dmg.sample(len(df_dmg))
    df_array = np.array_split(samples, chunks)
    #print("\n ===== chunks {} , df_size {}".format(chunks, len(df_array)))
    batch_path = os.path.join(batch_root_dir, workers[idx%5])
    worker_batches = np.array_split(df_array,5)
    if not os.path.exists(batch_path):        
        os.makedirs(batch_path)
    if workers[idx%5] not in damage_dict_batch_csv.keys():
        damage_dict_batch_csv[workers[idx%5]] = dict()
    else:
        pass
    batch = 0

    for worker_batch , worker in zip(worker_batches,workers):
        chunk_counter = 0
        for w_b in worker_batch:
            
            batch += 1
            csv_file_path = os.path.join(batch_root_dir,worker,'csv_data')
            csv_file_chunk = os.path.join(csv_file_path,"{}_batch_{}_of_{}.csv".format(damage, batch, chunks))
            if not os.path.exists(csv_file_path):        
                os.makedirs(csv_file_path)
            if not os.path.exists( os.path.join(ann_root_dit,worker,'data','annotation')):
                os.makedirs( os.path.join(ann_root_dit,worker,'data','annotation'))
            img_root = os.path.join(ann_root_dit,worker,'image')
            if not os.path.exists( os.path.join(ann_root_dit,worker,'image')):
                os.makedirs( os.path.join(ann_root_dit,worker,'image'))
            chunk_counter+=1
            w_b = pd.DataFrame(data=w_b, columns=['Unnamed: 0',
                     'Gutachter', 'Akten-ID', 'Damage-ID', 'Image', 'Location', 'Specific Location', 'Detailed Location',
                     'Damage', 'Repair Action', 'Cost', 'Undervalue', 'Description', 'status', 'dirty',
                     'action', 'damage', 'uid', 'part', 'part_p1', 'part_p2', 'bildname',
                     'damage_en', 'action_en', 'part_en',
                     'part_p1_en', 'part_p2_en'])

            w_b.drop(['Unnamed: 0'],axis=1,inplace=True)
            w_b['img_file_path'] = w_b.bildname.map(lambda x : os.path.join('./person',worker,'image',x))
            w_b['ann_file_path'] = w_b.bildname.map(lambda x : os.path.join('./person',worker,'data/annotation',x.split('.')[0]+'.xml'))

            try:
                w_b.to_csv(csv_file_chunk, index=False, sep=',')
            except:
                w_b.to_csv(csv_file_chunk.replace(' / ','-'), index=False, sep=',')

            if worker not in damage_dict_batch_csv.keys():
                damage_dict_batch_csv[worker] = dict()
            else:
                temp = dict()
                temp[csv_file_chunk] = dict()
                temp[csv_file_chunk]['Total Count'] = len(w_b)
                temp[csv_file_chunk]['Done Count'] = 0
                temp[csv_file_chunk]['Difficult Count'] = 0
                temp[csv_file_chunk]['Remain Count'] = len(w_b)
                #damage_dict_batch_csv[worker] = dict()
                damage_dict_batch_csv[worker].update(temp)
                

    
    idx+=1
```
